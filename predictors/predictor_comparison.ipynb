# Compare the performance of the different predictor models

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
attempts = ['CPM, ADAS11, Equal, Pos Network', 'CPM, ADAS11, Equal, Neg Network', 'CPM, ADAS11, Stratified, Neg Network', 'CPM, VentrVol, Pos Network', 'CPM, VentrVol, Neg Network', 'GB Regressor, VentrVol', 'GB Regressor, VentrVol with PCA', 
             'GB Regressor, VentrVol with ABeta and Tau', 'GB Regressor, ADAS11', 'RandomForrest, PHC_MEM', 'SVM, PHC_MEM', 'SVM, PHC_EXF', 'SVM, PHC_LAN', 
             'SVM, PHC_VSP', 'CNN, PHC_MEM']

values = [0.0026, 0.0020, -0.0881, 0.1104, -0.1224, 0.2035, -0.1041, -0.1658, 0.0658, -0.06, 0.07, 0.04, -0.02, -0.05, -0.5954]

colors = plt.cm.viridis(np.linspace(0, 1, len(attempts)))

plt.figure(figsize=(12, 8))

plt.bar(attempts, values, color=colors)

plt.title('R2 Score on Test Set across Different Model Trials', fontsize=16)
plt.xlabel('Model Variations Trialled', fontsize=14)
plt.ylabel('R2 Score', fontsize=14)

plt.xticks(rotation=45, ha='right', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

for idx in [5, 9, 10, 14]:
    plt.axvline(x=idx - 0.5, color='gray', linestyle='--')

plt.show()
```
